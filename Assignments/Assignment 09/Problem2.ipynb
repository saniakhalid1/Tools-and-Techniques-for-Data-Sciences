


### Analyze the distribution of movie genres over time.




```python
imdb_df=imdb_df.sort_values(by='IMDB_Rating',ascending=False)
top_movies = imdb_df[['Series_Title','Genre','IMDB_Rating','Gross']].head(20)
print(top_movies)
genre_counts = top_movies['Genre'].value_counts()
top_genres = genre_counts.head(10)
```

```python
plt.figure(figsize=(10,6))
top_genres.plot(kind='bar',color='skyblue')
plt.title('Top Genres in Top 100 movies with highest IMDB Ratings')
plt.xlabel('Genre')
plt.ylabel('Number of movies')
plt.xticks(rotation = 45,ha ='right')
plt.show()
```

```python
imdb_df['Released_Year'] = pd.to_datetime(imdb_df['Released_Year'], errors='coerce')
genres = imdb_df['Genre'].str.split(', ', expand=True).stack().unique()
genre_df = pd.get_dummies(imdb_df['Genre'].str.split(', ', expand=True).apply(lambda x: x.str.strip()), prefix='', prefix_sep='').groupby(level=0, axis=1).sum()
imdb_df = pd.concat([imdb_df, genre_df], axis=1)
genre_by_year = imdb_df.groupby(imdb_df['Released_Year'].dt.year)[genres].sum()
plt.figure(figsize=(14, 8))
genre_by_year.plot(kind='area', stacked=True)
plt.title('Distribution of Movie Genres Over Time')
sns.set_palette("husl") 
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```




### Identify the top-rated directors and actors.




```python
top_directors = imdb_df.groupby('Director')['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
print("Top-rated Directors:\n", top_directors)
actors_df = imdb_df[['Star1', 'Star2']]
actors = pd.concat([actors_df['Star1'], actors_df['Star2']]).dropna()
top_actors = actors.value_counts().head(10)
print("\nTop-rated Actors:\n", top_actors)
```




### Explore the correlation between budget, revenue, and ratings.




```python
imdb_df['Gross'] = pd.to_numeric(imdb_df['Gross'].str.replace(',', ''), errors='coerce')
correlation_matrix = imdb_df[['Gross', 'IMDB_Rating']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Between Revenue and Ratings')
plt.show()
```




### Utilize Pandas for advanced indexing and filtering




```python
filtered_movies = imdb_df[(imdb_df['Released_Year'] >= '2000-01-01') & (imdb_df['IMDB_Rating'] >= 8.0) & (imdb_df['Gross'] > 0)]
selected_columns = ['Series_Title', 'Released_Year', 'IMDB_Rating', 'Gross', 'Genre', 'Director']
filtered_movies = filtered_movies[selected_columns].sort_values(by='IMDB_Rating', ascending=False).head(10)
print("Top-rated movies since 2000 with Revenue > 0:\n", filtered_movies)
```
